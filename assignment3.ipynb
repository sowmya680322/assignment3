4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

````python
#variation for maths based on gender
sns.barplot(x = 'gender',y= 'math score'  ,data=data)
plt.show()
#variation for reading score based on gender
sns.barplot(x = 'gender',y= 'reading score'  ,data=data,color="yellow")
plt.show()
#variation for writing score based on gender
sns.barplot(x = 'gender',y= 'writing score' ,data=data,color="green")
plt.show()
````

````python
#variation for score ,reading score writing score based on gender
sns.barplot(x = 'test preparation course',y= 'writing score' ,data=data,color="green")
plt.show()
sns.barplot(x = 'test preparation course',y= 'reading score' ,data=data,color="pink")
plt.show()
sns.barplot(x = 'test preparation course',y= 'math score' ,data=data,color="red")
plt.show()
````
